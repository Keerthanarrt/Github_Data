```python
a=0
b=0
while b<3:
  while a<10:     #-->this is the nearest loop
    print(a)
    if a==3:
      break        # break will stop only the nearest loop and rest of the loops will contiued
    a+=1
  b+=1
```

```python
#continue
a=0
while a<10:
  print(a)
  if a==5:
    continue
  a+=1
```

```python
#pass type 1(indentation)
a=0
while a<10:
  a+=1
  if a==5:
    pass
  print(a)           #pass is used in if condition when there is nothing to perform aft if (without pass it will give syntax error)
```
